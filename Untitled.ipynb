```python
from wordcloud_fa import WordCloudFa
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
wr_img=cv2.imread('wordimg.png',3)
mask=cv2.imread('mask.png')
org_img= cv2.imread('org_img.jpg')
```

```python

plt.imshow(wr_img,cmap='gray')
plt.show()
```

```python
ret,trh_img=cv2.threshold(wr_img,100,200,cv2.THRESH_BINARY)
```

```python
plt.imshow(trh_img)
plt.show()
```

```python
mix_img=mask|trh_img
```

```python
plt.imshow(mix_img,cmap='gray')
plt.show()
```

```python
inv_maks=255-mask
```

```python
plt.imshow(inv_maks,cmap='gray')
plt.show()
```

```python
mix_img2=inv_maks|org_img
```

```python
plt.imshow(mix_img2,cmap='gray')
plt.show()
```

```python
mix_img3=mix_img&mix_img2
plt.imshow(mix_img3,cmap='gray')
plt.show()
```

```python
cv2.imwrite("./f.png",mix_img3)
```
