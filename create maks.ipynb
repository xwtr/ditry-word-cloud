```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
org_img= cv2.imread('org_img.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(org_img)
plt.show()

```

```python
ret,trh_img=cv2.threshold(org_img,7,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(trh_img,cmap='gray')
plt.show()
```

```python
kernel = np.ones((3,3),np.uint8)
closing_img = cv2.morphologyEx(trh_img, cv2.MORPH_CLOSE, kernel)
```

```python
cv2.imwrite("./mask.png",closing_img)
```

```python
plt.imshow(closing_img,cmap='gray')
plt.show()
```
